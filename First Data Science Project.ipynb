23. Use the codes for each country to merge the files so that you can convert the salary data to Australian Dollars (AUD). Print out the maximum and median salary in AUD. Hint: think about what data type you have.

```python
# Your code
#merging the dataframes based on CompensationCurrency column
rates.rename(columns = {'originCountry':'CompensationCurrency'}, inplace= True)  # renaming the column with CompensationCurrency so that it matches with first dataframe coulmn 
country_rates = pd.merge(responses, rates, how="left", on="CompensationCurrency")  # left merging 
country_rates['Salary in AUD'] = country_rates['CompensationAmount'] * country_rates['exchangeRateAUS']  # Conversion of Salary of respondents in AUD 
print('Maximum salary of respondents in AUD:',country_rates['Salary in AUD'].max())  #  Calculating Maximum Salary of respondents in AUD
print('Median salary of respndents in AUD:',country_rates['Salary in AUD'].median())  # Calculating median salary of respondenst in AUD
 
```
